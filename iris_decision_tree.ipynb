```python
from ucimlrepo import fetch_ucirepo,list_available_datasets
```

```python
#print the avialable UCI ML datasets
list_available_datasets()
```

```python
# import the data
iris_dataset = fetch_ucirepo(id=53)
iris_dataset.metadata
```

```python
#splitting of data using library or manually
from sklearn.model_selection import train_test_split
x=iris_dataset.data.features
y=iris_dataset.data.targets
x_train,x_test,y_train,y_test = train_test_split(iris_dataset.data.features,iris_dataset.data.targets,test_size=0.25,random_state=42)
print(f"{x_train.shape=}{x_test.shape=}")
```

```python
# create a decision tree
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,classification_report
import matplotlib.pyplot as plt
decision_tree = DecisionTreeClassifier()
decision_tree.fit(x_train,y_train)
```

```python
y_pred = decision_tree.predict(x_test)
print(f"{accuracy_score(y_test,y_pred)=}")
```

```python
#preview the decision tree
plt.figure(figsize=(10,10))
plot_tree(decision_tree=decision_tree,\
          class_names=y["class"].unique(),feature_names=x.columns)
plt.show()
```
